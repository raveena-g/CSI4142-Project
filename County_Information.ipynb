```python
#Defining path to census 2000-2010 files to be able to append all of them into one data frame
import glob
```

```python
import os
```

```python
path = '/content/Census_data/2000-2010'
```

```python
extension = 'csv'
```

```python
os.chdir(path)
```

```python
result = glob.glob('*.{}'.format(extension))
```

```python
print(len(result))
```

```python
import pandas as pd
```

```python
#Appending all census data from 2000-2010 together
df_csv_append = pd.DataFrame()
for file in result:
    print(file)
    df = pd.read_csv(file, encoding='latin-1')
    df_csv_append = df_csv_append.append(df, ignore_index=True)
 
df_csv_append
```

```python
#Year is not defined as an absolute year, but a code that refers to a certain year. Certain year codes need to be dropped, given value '0'
def categorise(row):
    if row['YEAR'] == 1:
      return 0 
    if row['YEAR'] == 2:
        return 2000
    if row['YEAR'] == 3:
        return 2001
    if row['YEAR'] == 4:
        return 2002
    if row['YEAR'] == 5:
        return 2003
    if row['YEAR'] == 6:
        return 2004
    if row['YEAR'] == 7:
        return 2005
    if row['YEAR'] == 8:
        return 2006
    if row['YEAR'] == 9:
        return 2007
    if row['YEAR'] == 10:
        return 2008
    if row['YEAR'] == 11:
        return 2009
    if row['YEAR'] == 12:
        return 0
    if row['YEAR'] == 13:
        return 2010    
    return 0
```

```python
#New year column "POP_YEAR" is created
df_csv_append['POP_YEAR'] = df_csv_append.apply (lambda row: categorise(row), axis=1)
```

```python
df_csv_append
```

```python
#Age group 0 defines whole population, which is what we are looking at, so every other age group is dropped
df_csv_append = df_csv_append[df_csv_append.AGEGRP == 0]
```

```python
df_csv_append
```

```python
#FIPS code is given as county and state seperately
#New code has to be created which is a composite of both so that other data can be joined later on
#Creating a string version of the county column
df_csv_append['FIPS_COUNTY'] = df_csv_append['COUNTY'].astype(str)
```

```python
#Adding leading zeroes to match necessary format
df_csv_append['FIPS_COUNTY'] = df_csv_append['FIPS_COUNTY'].str.zfill(3)
```

```python
#Creating string version of state column
df_csv_append['FIPS_STATE'] = df_csv_append['STATE'].astype(str)
```

```python
#Adding leading zeroes to match necessary format
df_csv_append['FIPS_STATE'] = df_csv_append['FIPS_STATE'].str.zfill(2)
```

```python
#Creating FIPS code which is a composite of state and county
df_csv_append['FIPS_CODE'] = df_csv_append['FIPS_STATE'] + df_csv_append['FIPS_COUNTY']
```

```python
df_csv_append
```

```python
#Dropping years that we do not want
df_csv_append = df_csv_append[df_csv_append.POP_YEAR != 0]
```

```python
df_csv_append
```

```python
#Downloading csv
from google.colab import files
df_csv_append.to_csv('Census2000-2010.csv', encoding = 'utf-8-sig') 
files.download('Census2000-2010.csv')
```

```python
#Repeating all steps for data from 2010-2020
path = '/content/Census_data/2010-2020'
extension = 'csv'
os.chdir(path)
result = glob.glob('*.{}'.format(extension))
```

```python
#Defining the year categories based on new dataset
def categorise2010(row):
    if row['YEAR'] == 1:
      return 0 
    if row['YEAR'] == 2:
        return 0
    if row['YEAR'] == 3:
        return 0
    if row['YEAR'] == 4:
        return 2011
    if row['YEAR'] == 5:
        return 2012
    if row['YEAR'] == 6:
        return 2013
    if row['YEAR'] == 7:
        return 2014
    if row['YEAR'] == 8:
        return 2015
    if row['YEAR'] == 9:
        return 2016
    if row['YEAR'] == 10:
        return 2017
    if row['YEAR'] == 11:
        return 2018
    if row['YEAR'] == 12:
        return 2019
    if row['YEAR'] == 13:
        return 2020    
    return 0
```

```python
#Applying new year function
df_csv_append2020['POP_YEAR'] = df_csv_append2020.apply (lambda row: categorise2010(row), axis=1)
df_csv_append2020
```

```python
#Dropping unnecessary age groups
df_csv_append2020 = df_csv_append2020[df_csv_append2020.AGEGRP == 0]
```

```python
#Creating composite FIPS key
df_csv_append2020['FIPS_COUNTY'] = df_csv_append2020['COUNTY'].astype(str)
df_csv_append2020['FIPS_COUNTY'] = df_csv_append2020['FIPS_COUNTY'].str.zfill(3)
df_csv_append2020['FIPS_STATE'] = df_csv_append2020['STATE'].astype(str)
df_csv_append2020['FIPS_STATE'] = df_csv_append2020['FIPS_STATE'].str.zfill(2)
df_csv_append2020['FIPS_CODE'] = df_csv_append2020['FIPS_STATE'] + df_csv_append2020['FIPS_COUNTY']
df_csv_append2020
```

```python
#Repeating steps for 2021 data
path = '/content/Census_data/2020-2021'
extension = 'csv'
os.chdir(path)
result = glob.glob('*.{}'.format(extension))
df_csv_append2021 = pd.DataFrame()
for file in result:
    print(file)
    df = pd.read_csv(file, encoding='latin-1')
    df_csv_append2021 = df_csv_append2021.append(df, ignore_index=True)
 
df_csv_append2021
```

```python
#Creating one data frame which is all census data from 2000-2021
df_census_data = pd.DataFrame()
df_census_data = df_census_data.append(df_csv_append, ignore_index=True)
df_census_data = df_census_data.append(df_csv_append2020, ignore_index=True)
df_census_data = df_census_data.append(df_csv_append2021, ignore_index=True)
```

```python
#Reading in health data 
health_data = pd.read_csv("/content/Health_data/2010-2022_Health_Data.csv")
health_data.head()
```

```python
#Creating new FIPS column which is string to be able to join with census data
health_data['FIPS_NEW'] = health_data['FIPS'].astype(str)
health_data.head()
```

```python
#Adding leading zeroes necessary for format
health_data['FIPS_NEW'] = health_data['FIPS_NEW'].str.zfill(5)
```

```python
health_data.head()
```

```python
#Joining census and health data on FIPS and year
census_health = pd.merge(df_census_data, health_data,  how='left', left_on=['FIPS_CODE','POP_YEAR'], right_on = ['FIPS_NEW','Year'])
census_health.head()
```

```python
#Creating surrogate keys
census_health = census_health.assign(County_Key=range(len(census_health)))
census_health
```

```python
census_health=census_health.drop(['FIPS', 'State', 'County', 'Year', 'FIPS_NEW','SUMLEV', 'YEAR', 'AGEGRP', 'STATE', 'COUNTY'], axis=1)
census_health
```
