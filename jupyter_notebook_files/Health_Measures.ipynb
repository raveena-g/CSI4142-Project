```python
import pandas as pd
```

```python
#Reading in health data 
health_data = pd.read_csv("/content/Health_data/2010-2022_Health_Data.csv")
health_data.head()
```

```python
#Creating new FIPS column which is string to be able to join with other data
health_data['FIPS_NEW'] = health_data['FIPS'].astype(str)
#Adding leading zeroes necessary for format
health_data['FIPS_NEW'] = health_data['FIPS_NEW'].str.zfill(5)
#Dividing 5 character FIPS code into 2 codes for State and County
code_elements = ['state_code', 'county_fips']
health_data[code_elements] = health_data['FIPS_NEW'].str.extract(r'^(\d{2})(\d{3})')
```

```python
health_data.head()
```

```python
health_data = health_data.drop('FIPS', axis = 1)
```
